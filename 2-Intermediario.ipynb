```python
import pandas as pd
from datetime import datetime
```

```python
# Retona numero de linha e colunas
def get_shape(df) -> None:
    linha, coluna = df.shape
    print(f"""Número de linhas: {linha}\nNúmero de colunas: {coluna}""")
```

<p style="font-family: Arial; font-size:1.75em;color:#004AAD; font-style:bold">lambda</p>
<p> 
    Em vez de criar uma função nomeada, podemos criar uma função lambda (função anônima) diretamente no código.<br>
    <b>lambda parameters: expression</b>
</p>

```python
# Função tradiconal
def func_tradicional(x):
    x = x + 10
    return x


# Função lambda
func_lambda = lambda x: x+10

print("Função tradicional: ",func_tradicional(2))
print("Função lambda: ",func_lambda(2))
```

<p style="font-family: Arial; font-size:1.75em;color:#004AAD; font-style:bold">pandas.transform</p>
<p> 
    Aplica uma transformação (função) em todos os elementos do data frame
</p>

```python
# Data frame básico
df = pd.DataFrame({"A":[1, 2, 3],  
                   "B":[4, 5, 6]}) 
df
```

```python
df.transform(lambda x: x+10)
```

```python
df['B'] = df[['B']].transform(lambda x: x+10)
df
```

<p style="font-family: Arial; font-size:1.75em;color:#004AAD; font-style:bold">Agrupamento</p>

```python
# Soma dos valores dos módulos por cliente
valores_plano_cliente = line_item[['deal_id', 'Valor Módulo']].groupby('deal_id').sum()
valores_plano_cliente
```

```python
resumo = pd.merge(cliente,valores_plano_cliente, how='inner', on='deal_id')
resumo.rename(columns={'Valor Módulo': 'Total Contratado'}, inplace=True)
resumo = resumo.drop(columns=['deal_id'])
resumo
```

```python
# Valor médio por tipo de negócio
valor_medio_tn = resumo[['Tipo de Negócio',
                         'Total Contratado'
                         ]].groupby('Tipo de Negócio').mean()

valor_medio_tn.rename(columns={'Total Contratado':
                                'Valor médio'}, inplace=True)

valor_medio_tn
```

<p style="font-family: Arial; font-size:1.75em;color:#004AAD; font-style:bold">Agrupamento com multiplas agregações</p>

```python
# Valor máximo por tipo de negócio
agregacoes = resumo[['Tipo de Negócio',
                         'Total Contratado'
                       ]].groupby('Tipo de Negócio').agg(['count','mean','min','max'])

agregacoes.rename(columns={'Total Contratado':
                           'Agregações'}, inplace=True)

agregacoes
```
